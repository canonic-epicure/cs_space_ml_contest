```python
import pandas as pd
import numpy as np
```

## Read data

```python
df = pd.read_csv('history.csv')
```

```python
df
```

## Create submission

```python
df['is_active'] = 0
df['next_month_revenue'] = 0.0
```

```python
df_sorted = df.sort_values('id', ascending=True)
```

```python
df_unique = df_sorted.drop_duplicates(subset='id', keep='first')
submission = df_unique[['id', 'next_month_revenue', 'is_active']]
```

```python
submission
```

## Save submission

```python
submission.to_csv('submission.csv', index=False)
```
